#### When to use - 

Objective: Model $P(x)$ for density estimation (which often use EM) - 

* For: anomoly detection.
* Use: Probabilistic analysis (e.g., compute P(x)).
* Data in subspaces (e.g., high dimentional, but actually lies in lower dimentional sub-space): Factor analysis.
* Data in clusters (e.g., clumps): Mixture of Gaussians.

Objective: Non-probablistic model for data compression - 

* For: Visualization or understanding.
* Use: Non-probabilistic analysis.
* Data in subspaces: PCA.
* Data in clusters: k-means.

#### Distance metrics - 

(1) Euclidian distance (between vectors - two lists of numbers): 
$$ \sqrt{\sum (x_i - y_i)^2} $$ 

(2) Correlation (of two lists of numbers):
$$ \frac{cov(x,y)}{\sigma_x \sigma_y} $$ 

(3) Manhattan (k=1) and Cheychev (k=inf) and k-th root:
$$ \sqrt{\sum (x_i - y_i)^k} $$ 

(4) Angle
$$ acos \frac{abs(x \cdot y)}{abs(x) abs(y)} $$ 

#### K-means - 

This algorithm is intiutive:

* Assign each training example, $x^i$, to a cluster centroid, $\mu_j$.
* Moving the cluster centroid ($\mu_j$) to the mean of points assigned to it.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
%matplotlib inline
import sklearn.datasets as datasets
# Make data with four clusters (X are data, Y are centers)
X,Y = datasets.make_blobs(centers=3, cluster_std=0.5, random_state=0)
plt.subplot(1,2,1)
plt.jet() # Set the color map.
plt.scatter(X[:,0], X[:,1],c=Y)
plt.xticks(fontsize=5)
plt.yticks(fontsize=5)
plt.grid()
plt.title('Pre-clustered data',fontsize=7.5)

plt.subplot(1,2,2)
from sklearn.cluster import KMeans
kmeans = KMeans(3, random_state=8) # Random state is the generator used to initialize the centers
Y_hat = kmeans.fit(X).labels_
plt.scatter(X[:,0], X[:,1], c=Y_hat, alpha=0.4)
mu = kmeans.cluster_centers_
plt.scatter(mu[:,0], mu[:,1], s=100, c=np.unique(Y_hat))
plt.xticks(fontsize=5)
plt.yticks(fontsize=5)
plt.grid()
plt.title('Assign clusters with kMeans',fontsize=7.5)
```

#### Hierarchical clustering -

I just use `Seaborn` for this.

It is very easy to plot hierarchically clustered heatmap of a pandas DataFrame.

```python
flights = sns.load_dataset("flights")
flights.head(3)
```

Let's pivot the data.

```python
flights_rect = flights.pivot("month", "year", "passengers")
flights_rect.head(3)
```

Then, we can cluster it.

```python
cg = sns.clustermap(flights_rect)
```

This is a little skewed because the number of flights increase by year.

So let’s standard_scale the data (i.e. divide all the columns by the maximum so we can compare year-to-year on the same scale). 

```python
cg = sns.clustermap(flights_rect, standard_scale=1) # Scale columns!
```

We can also normalize by Z-score:

* Subtracts the mean 
* Divides by the standard deviation

So, each column will have 0 mean and a variance of 1. 

Each value will simply indicate the number of standard deviations away from the mean! 

```python
cg = sns.clustermap(flights_rect,z_score=1) # Scale columns!
```
