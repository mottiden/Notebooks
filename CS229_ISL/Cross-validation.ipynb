#### Bias and variance - 

Generalization error occours when our model does not generalize well to data outside the training set. 

For regression, we can use mean squared error (MSE) to evaluate the performance of a learning method on a given data set.
 
$$ MSE = \frac{1}{n} \sum\limits_{i=1}^n (y^i-\hat{f}(x_i))^2 $$ 
 
The MSE can always be decomposed into the sum of three fundamental quantities:

$$ MSE = (y_o - \hat{f}(x_o))^2 = B^2 + V + var(\epsilon)$$

Bias, $B$: 

* Error introduced by using a model that is simple relative to actual relationship (e.g., under-fitting). 
* Generalization error even if we were to fit it to a very (e.g., infinitely) large training set. 
* For example, a linear model fit to non-linear data (e.g., underfit) has high bias.

Variance, $V$: 

* Amount by which $\hat{f}$ would change if we estimated it using a different training data set (e.g., over-fitting).
* Generalization error if we fit “spurious” patterns in the training set. 
* For example, a high order polynomial fit to the noise in linear data (e.g., overfit) has high variance.

In order to minimize the expected test error, we need to select a statistical learning method that simultaneously achieves low variance and low bias. 

Prectically, there is often a trade-off between these can be observed: 

* A complex model with many parameters may have low Bias, but high Variance.


#### The case of cross-validation - 

Here's a naive strategy for model testing:

* Train each model on a training set, to get some hypothesis $h_i$.
* Pick the hypotheses with the smallest training error.

This fails!

For example, a high-order polynomial can easily fit a training set well. But it does not generalize beyond it!

It has high variance, and overfits.

Thus, a better idea is to split up the data that we train and test on using cross-validation.

#### Cross-validation - 

Validation set - 

We randomly dividing the available set of observations into two parts, a training set and a validation set.

Drawbacks: 
    
* Variance: The estimate of the test error rate can be highly variable, depending on observations in the training set.
* Test error: Only a subset of the observations are used to fit the model.
        
---

Leave-one-out cross-validation (LOOCV) - 

The learning method is fit on $n − 1$ training observations.

A prediction $yˆ1$ is made for the excluded observation, using its value $x_1$.

This is repeated $n$ times, resulting in $n$ squared errors. 

The LOOCV estimate for the test MSE is the average of these n test error estimates:

Advantages over validation set:
        
* Less bias: We use almost as many as observations as are in the entire data set.
    
Drawback:
    
* LOOCV has the potential to be expensive to implement, since the model has to be fit $n$ times. 
    
* We train the model on an almost identical set of observations.

* Higher variance: Higher likelihood of over-fitting. The outputs are highly (positively) correlated with each other. 
    
---

K-folds -

This approach involves randomly dividing the set of observations into k groups, or folds, of approximately equal size. 

Advantages over LOOCV:

* It is less expensive, as it requires fitting the statistical learning method fewer times. 

* Lower variance: Averaging the outputs of $k$ fitted models that are somewhat less correlated with each other.

Drawback: 
    
* Intermediate level of bias, since each training set contains $\frac{(k − 1)n}{k}$ observations.

The test error estimate resulting from LOOCV tends to have higher variance than does the test error estimate resulting from k-fold CV.


#### Implementation - 

An interesting case-study on cross-validation and exparimental design:

https://followthedata.wordpress.com/2013/10/30/the-importance-of-proper-cross-validation-and-experimental-design/
    
A study claimed that a genetic test could be used to predict autism risk from birth based on a small number (237) of SNPs.

This was done using $5,000$ individuals with ASDs in a case/pseudo-control design, as well as a $4,500$ unrelated case/control design.

Problem 1 - Differences due to disease cannot be disentangled from technology- or ethnicity specific biases.
    
*  Batch effects: Case and control groups were characterized at different times and on different technological platforms.
*  Case and control groups had different ethnic composition.

Problem 2 - Feature selection:
    
* The authors had first performed feature selection on the whole data set, and then divided the data set into a training set and a validation set to assess the performance.
* This “contaminates” the model with information about the test set and can easily lead to poor generalization.


#### Cross-validation with feature selection - 

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
def makeTarget(x,f1,f2):
    # Make target a funtion of two features
    y=x[:,f1]+x[:,f2] # Simple linear combination of two columns
    y+=np.random.normal(0,.01,y.shape) # Add noise
    return y

# Make features
samples=30
x=np.random.uniform(0,10,(samples,10000))

# Generate target as a linear combination of f1, f2
f1,f2=0,1
y=makeTarget(x,f1,f2)
```

We want to find features that correlate with the target, making them reasonable candidates for inclusion in a model. In turn, we can use `SelectKBest` along with `f_regression`: the latter performs linear regression and outputs a score indicating quality of fit relative to the other features (F-test, see below) and then the former simply selects a specified number of highest scoring features. See http://en.wikipedia.org/wiki/F-test#Regression_problems for model score.
    

```python
from sklearn.feature_selection import SelectKBest,f_regression
selector = SelectKBest(f_regression,k=2).fit(x, y)
# Get a mask, or integer index, of the features selected
best_features = np.where(selector.get_support())[0]
print(best_features)
```

```python
feat=pd.DataFrame(x)
feat['target']=y
feat.columns=[str(c) for c in feat.columns] # Bug in seaborn: only works on string values in columns.
```

```python
sns.jointplot(str(f1),"target",feat,kind='scatter')
```

```python
sns.jointplot(str(f2),"target",feat,kind='scatter')
```

The target is a linear combination of features 1,2. 

But, we added noise. So, we identified a spurious correlations (below).

```python
sns.jointplot(str(best_features[0]),"target",feat,kind='scatter')
```

```python
sns.jointplot(str(best_features[1]),"target",feat,kind='scatter')
```

We use these two features to build our model.

```python
from sklearn.linear_model import LinearRegression
x=np.array(feat[best_features])
y=np.array(feat["target"])
clf = LinearRegression().fit(x,y)
print("Score is ",clf.score(x,y))
```

We then run with a specified number of $k$ folds. As mentioned, we split the data into equal partitions, train, and test.

```python
from sklearn.cross_validation import cross_val_score, KFold
cross_val_score(clf,x,y,cv=3).mean()
```

But, the score is similar. Seems OK!?

```python
for train, test in KFold(len(y),3):
    xtrain,xtest,ytrain,ytest = x[train],x[test],y[train],y[test]
    clf.fit(xtrain, ytrain) # Fit on training data
    yp = clf.predict(xtest) # Test data
    plt.scatter(yp,ytest,color="k",edgecolor="white",s=50,lw=1)
    plt.plot(ytest,ytest,'k',lw=0.5)
plt.xlabel("Predicted")
plt.ylabel("Observed")
```

Does this generalize? Let's test by making new data.

```python
x_new=np.random.uniform(0,10,(samples,10000))
y_new=makeTarget(x_new,f1,f2)
```

```python
yp=clf.predict(x_new[:,best_features]) 
plt.scatter(yp,y_new,color="k",edgecolor="white",s=50,lw=1)
plt.plot(y_new,y_new,'k',lw=0.5)
plt.xlabel("Predicted")
plt.ylabel("Observed")
```

We over-fit to the training data! The model doesn't generalize well to the new data generated.

This is because data is generated only from features 1,2.

But, the model was trained on a feature ($9357$) that randomly correlated with the target only for that dataset.

We did feature pruning *prior* to $k$-fold validation, so we never picked up this problematic over-fitting.

$\textbf{Rather, we want to perform feature pruning at every iteration of cross-validation.}$

In general, it is useful to understand why an algorithm works.

One approach is ablative analysis:
    
* Systematic removal of elements.
* Evaluate performance at each step.

Furthermore, performance or classification accurancy should be provided in context.

For example, we can have problems with highly unbalanced classes:

* 0.4% of people in the US have AIDS.
* Thus, a test that always says "no" will be 99.6% accurate.

One way to address this is a confusion matrix, which is the predicted versus the actual class labels:

* TP - True positive.
* FN - False negative.
* FP - False positive result.
* TN - True negative.

A simple way to summarize this table is the ROC curve:

* Precision - False positive rate. Of all the positives tagged, how many were true?
* Recall - What fraction of actual positive were correctly identified? 

ROC curve is a nice mapping of the relationship between precision and recall.

We use a confusion matrix to show a cross tabulation of the actual and predicted class labels: correct classifications are shown in the diagonal, the off-diagonal terms show classification errors (false-positive and false-negative). A condensed summary of performance is given by the misclassification rate (number of errors by the total number of cases).

```python
from sklearn.cross_validation import train_test_split
a_train, a_test, b_train, b_test = train_test_split(feat,targ,test_size=0.25)
clf.fit(a_train,b_train)
predicted=clf.predict(a_test)
# Accuracy
b_test=b_test.T
c=b_test[b_test==predicted]
print c.shape[0]/float(predicted.shape[0])
```

```python
from sklearn.metrics import confusion_matrix as sk_confusion_matrix
b_test=b_test.T
cm = sk_confusion_matrix(b_test,predicted)
cm = pd.DataFrame(data=cm, columns=['F','T'], index=['F','T'])
cm.columns.name = 'Model Label'
cm.index.name = 'Label'
cm
```

#### Model de-bugging - 

Scenario:

* Bayesian logistic regression has 20% test error.
* Using 100 features (words).
    
Many common approaches:

* Change size of feature set.
* Change different algorithm.

Usually one picks a method at random, which is not efficient.

Better approach is to use diagnostics to understand why the algorithm is not working.

Overfitting:

* High variance.
* Fitting high order polynomial to linear data.
* In this case, training error will be much lower than test error. 

Underfitting:

* High bias.
* Fitting low order polynomial to complex data.
* In this case, training error will also be high. 

To examine this, use a learning curve:

* Plot both test and training error with respect to the training set size.
* The test set error will usually go down as we increace the training set size.
* The training error usually goes up with respect to m, the training set size.
* It becomes harder to fit the model to more training examples well.

Results:

* Overfitting: A large gap between test and training is indicative of overfitting. To resolve, get more training data.
* Underfitting: Test and training error are similarly poor. So, here want to get a different model in order to better fit our data.
    
In summary, these approaches should be used to address high variance (overfitting):

* More training examples.
* Smaller set of features.

In summary, these approaches should be used to address high bias (underfitting):

* Larger set of features.

Objectives or cost functions:

* We might have a business or scientific objective.
* But, our algorithms have specific objectives (convex) functions. 
* We must ensure that this is a proper objective function to use.
