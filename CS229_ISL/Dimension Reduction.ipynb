#### Controlling variance - 

Recall that variance results from over-fitting a model to the training data!

This can be controlled variance in at least two different ways:
    
* Best subset selection: Using a subset of the original variables.
* Ridge regression: shrinking their coefficients toward zero. 

We can also transform the predictors and then fit a least squares model using the transformed variables.

#### PCA - 

Motivation:

* With high dimentional data, we might have redundant features.
* Compressing these can improve inference.
* As mentioned above, reducing features may also reduce variance (over-fitting problems).

We can define new features, which are just linear combinations of our initial $p$ predictors:

$$ Z_m = \sum_{i=1}^p \phi_{im} x_{i} $$

We can then fit a linear regression model to the $M$ combinations determined above:

$$ y_i = \theta_0 +  \sum_{i=1}^m \theta_{im} z_{i} + \epsilon $$

Simply, we now must learn $m$ rather than $p$ coefficients.

When $M < p$, this can significantly reduce the variance of the fitted coefficients.

The first principal component:
    
* Is the direction of the data is that along which the observations vary the most.
* Is the line that is as close as possible to the data (minimizes the sum of the squared perpendicular distances between each point and the line).

#### Pre-processing - 

Re-normalization rescales the attributes to make them more comparable:

* Compute the mean ($\mu = \frac{1}{m} \sum_{i=1}^m x_i$) and subject from each point ($x^i = x^i - \mu$) to get mean of zero.
* Compute the variance ($\sigma^2 = \frac{1}{m} \sum_{i=1}^m (x_i - \mu)^2$) and divide it by each point by $\sigma$ to get variance of one.


#### Algorithm - 

The principal component of the data:

* Let $u$ be the principal component, with norm of 1.
* The length of the projection of any point, $x^i$, onto $u$ is: $(x^i)^T u$.

Recall that the variance of a set of $n$ equally likely values can be written as:

$$ \frac{1}{n} \sum_{i=1}^n (x^i - \mu)^2 $$

Moreover, we will choose $u$ such that it maximizes the variance of the projections of the data onto $u$:

$$ max_{\| u \| = 1} \frac{1}{m} \sum_{i=1}^m ((x^i)^T u)^2 $$

This just maximizes the sum of the data mapped onto $u$ and can be re-written:

$$ max_{\| u \| = 1} u^T [\frac{1}{m} \sum_{i=1}^m x^i {x^i}^T] u $$

Note that one of our terms is simply the covariance matrix of our data:

$$ max_{\| u \| = 1} u^T \Sigma u $$

The solution to this problem is that $u$ is the principal eigenvector of the covariance matrix, $\Sigma$.

We also know that it is easy to solve for the eigenvector, $u$, and eigenvalue, $\lambda$:

$\Sigma u = \lambda u$

Thus, in order to find the dimention of greatest variation and project this data onto this one-dimentional sub-space, we can:

* Compute the covariance matrix.
* Compute the principal eigenvector.

Of course, we can also find the top $k$ eigenvectors:

* These are the vectors that satisfy $\Sigma u = \lambda u$.
* With the largest possible values of $\lambda$.

In sum, we can often capture a large fraction (e.g., 0.95) of the variance of our data (across all $n$ dimentions) with a small set of $k$ principal components:

$$ \frac{\sum_{i=1}^k \lambda_i}{\sum_{i=1}^n \lambda_i} = 0.95 $$

#### Implementation - 

Ok, here's the easy way to think about this.

We start with some data.

```python
import sys,os
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt 
%matplotlib inline

# Bi-variate Gaussian
u = np.array([0,0])
cov = np.array([[1,0.85],[0.85,1]])
X = np.random.multivariate_normal(u, cov, 10000)
sns.jointplot(X[:,0], X[:,1], kind="hex")
```

Recall covariance indicates the linear relationship between variables:

$$ Cov(a,b) = \frac{1}{n-1} \sum\limits_{i=1}^{n} (a_i - \mu_a) (b_i - \mu_b) $$

The off-diagonal of the covariance matrix indicates this relationship.

Larger values compress the data into a line.

```python
# We can compute this
print np.cov(X.T)
```

The eigenvectors (principal components) are:

* Orthogonal directions of greatest and least variation.
* Linear combinations of the original axes.

The eigenvalues:

* Equal to the variances of the data along the eigenvectors.

They can be computed by eigendecomposition on the covariance matrix.

```python
from numpy import linalg as LA
w,v=LA.eig(np.cov(X.T))
print 'The eigenvalues are %.2f and %.2f' % (w[0],w[1]) #eigenvalues
```

The eigenvectors.

```python
print v
```

We can overlay the eigenvectors on our data, showing the direction of greatest variance (yellow).

```python
plt.subplot(111,aspect='equal')
ah=0.1 
f=1.1
plt.arrow(0,0,v[0,0],v[1,0],color='r',linewidth=2,head_width=ah,head_length=ah)
plt.arrow(0,0,v[0,1],v[1,1],color='y',linewidth=2,head_width=ah,head_length=ah)
plt.scatter(X[:,0], X[:,1])
```

The features can have different distributions.

We normalize by z-score prior to applying PCA:

* Subtract the mean, giving each feature a mean value of zero.
* Divide by standard deviation, giving each feature $\sigma^2=1$.

See here:

http://stats.stackexchange.com/questions/69157/why-do-we-need-to-normalize-data-before-analysis

```python
X_1=np.subtract(X,X.mean(axis=0))
X_2=np.divide(X_1,X_1.std(axis=0))
sns.jointplot(X_2[:,0],X_2[:,1],xlim=(-5,5),ylim=(-4,4))
```

Now, let's apply PCA.

Basicallt, we rotate the data.

We can see that variance is maximized on `PC1` (x-axis).

```python
from sklearn.decomposition import PCA
n_comp = 2
pca = PCA(n_components=n_comp)
X_t = pca.fit(X_2).transform(X_2)
sns.jointplot(X_t[:,0],X_t[:,1],xlim=(-5,5),ylim=(-4,4))
```

The loadings are the eigenvectors.

```python
loadings = pca.components_
print loadings
```

We know that these specify the direction of the principal components.

We can show this below directly from the eigenvector.

```python
x=np.linspace(np.min(X_2[:,0]),np.max(X_2[:,0]),100)
m=loadings[0][0]/loadings[0][1]                 
y=m*x
plt.scatter(X_2[:,0],X_2[:,1])
plt.plot(x,y,c='k')
```

If the data were only varied along one dimention, then the eigenvector would point be a horizontal or vertical line.

One feature would have a eigenvector value of zero.

So, non-zero eigenvector values (loadings) contribute more strongly to defining the direction of the eigenvector.

As a result, we can use this as a feature pruning trick: non-zero loadings are feature that more strongly contribute to the variance in a dataset!
