##What is Python

**From the docs**

*https://docs.python.org/2/faq/general.html*

*Python is an interpreted, interactive, object-oriented programming language. It incorporates modules, exceptions, dynamic typing, very high level dynamic data types, and classes. Python combines remarkable power with very clear syntax. It has interfaces to many system calls and libraries, as well as to various window systems, and is extensible in C or C++. - interactive = can act as a shell - statements can be run one at a time

**TLDR;**

Python is a B+ language at everything, its extensibility and the focus from the community onto scientific libraries in the 90s is why it is used so much in data analysis.

## Other languages

*Given that Revolution Analytics just got bought by Microsoft I think its appropriate to discuss Python vs R*

**Python**
- Python is a programming language and is thus similar to other programming languages (R is known to be unintuitive to those who know other languages)
- Code is much more production ready than R - this is especially good if you have Python web developers
- With Machine Learning, Python is more focused in predictive accuracy (e.g. cross validation)
- There are a lot more things to think about when creating a model in Python such as data structures
- It is A LOT easier to swap between models in sk-learn than in R once you get a hang of it, where swapping models can mean swapping libraries (R's caret *(Classification And REgression Training)* is an answer to this)
- Data manipulation is a lot easier in Python - regular expressions are easier than R and Python has more data structures
- Natural Language Processing is a lot easier in Python thanks to NLTK

**Why R**
- R is MUCH more focused on data analysis
- Bleeding edge packages are generally easier to find in R
- Used much more widely in academia
- With Machine Learning, R is much more focused on statistical learning (i.e. easier to access and run diagnostics on models)
- Creating your first model is a lot easier in R
- Installing R and RStudio is A LOT easier than Python
- dplyr is said to be easier than Pandas (I'm not sure of this)
- ggplot is MILES ahead of matplotlib and other Python libraries

**Why Neither**
- Ther is a *lot* to learn to master both R and Python
- Can cause bottlenecks in organisations if only one person knows 
- Can end up as just another tool that only one person uses
    - i.e. Lisp
- Limited Support and Talent especially in Perth
- Not as much support as vendors (especially regarding integration) but debatably better documentation

## Installing Python

**Python 2 or 3?**

- Unless you KNOW RIGHT NOW that you will use a package that is only avaliable in 2.X, use 3

**Installing Python is a pain in the butt for a few reasons:**

- Some libraries are compiled, Windows hates these
- There are some best practices that not every tutorial will tell you, if you don't follow them life can become miserable
  - If your system already has Python, don't use that installation
  - Keep everything in virtual environments and use them
  
I explain how to set up anaconda here: https://github.com/jackgolding/FullStackDataAnalysis

## Virtual Environments

**Keep your Python Installations and Libraries Isolated**

When you update pandas from 0.14 to 0.15 for a new project, you don't want to break all your old projects!

**Create and activate a new environment in Anaconda**

conda info -e

conda create -n new_environment --clone root

source activate new_environment

use pyenv/virtualenv if you don't want to install Anaconda

## CRISP-DM
Cross Industry Standard Process for Data Mining

- Currently championed by IBM, created in mid 90s

- leading methodology for data mining (KDD, surveys 2002, 2004, 2007)
- Puts business understanding at the front and centre
- All other methologies I have seen are waterfall-esque

- Initial briefing, requirements
- Book meetings with stakeholders
- Try to independantly understand data and business rules

  - Pandas
