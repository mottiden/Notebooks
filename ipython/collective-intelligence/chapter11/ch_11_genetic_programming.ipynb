### 유전자 프로그래밍의 정의?
<b>다윈의 자연선택 이론(진화론)</b>에 기반을 둔 알고리즘이다.

1. 일반적으로 이는 무작위나 수작업으로 생성되어 다소 좋은 해답으로 간주되는 개체군(population)이나 큰 프로그램 집합을 가지고 시작
2. 그 후 이 프로그램들은 몇몇 사용자 정의 작업(경쟁)을 수행
3. 경쟁을 마치면 최적에서 최악까지 프로그램들의 순서를 결정
4. 최적에 가까운 프로그램들을 돌연변이 혹은 교배 연산 등을 통해 진화시켜 답을 찾아 나감

각 단계에서 프로그램의 품질은 <b>적응도 함수(fitness function)</b>을 통해 계산한다.

개체군의 크기는 항상 일정하게 유지되므로 <u>최악 프로그램들의 대부분은<br>
개체군에서 새로운 프로그램을 만들 공간을 확보하기 위해 제거된다.</u>

새롭게 생성된 개체군을 <b>다음세대(next generation)</b>이라 부르며, <b>전체 과정들이 반복</b>된다.

최적 프로그램은 항시 유지되고 변경되기 때문에 청소년 세대가 부모 세대보다 더 똑똑해지는 것과 같이 <u>매 세대마다 점점 좋아질 것</u>이다.

1. 완벽한 해답을 찾았다.
2. 충분히 훌륭한 해답을 찾았다.
3. 여러 세대에 걸쳐 해답이 향상되지 않았다.
4. 세대 수가 정해진 한계에 도달하였다.

## - 파이썬에서 트리 표현하기 -

아래는 진화지성을 테스트하기 위한 Parse Tree 코드이다.

```python
from random import random, randint, choice
from copy import deepcopy
from math import log

# 함수 노드에서 사용할 삼수들에 대한 래퍼.
# 멤버 변수로 함수 이름, 함수 자신, 취할 인자의 개수를 가짐
class fwrapper:
    def __init__(self, function, childcount, name):
        self.function = function
        self.childcount = childcount
        self.name = name

# 자식 노드를 가진 함수 노드 클래스.
# fwrapper로 초기화된다. evaluate가 호출되면 자식 노드들을 평가한 후 그 결과로 이 함수를 호출한다.
class node:
    def __init__(self, fw, children):
        self.function = fw.function
        self.name = fw.name
        self.children = children
        
    def evaluate(self, inp):
        results = [n.evaluate(inp) for n in self.children]
        return self.function(results)
    
# 프로그램에 전달된 인자들 중 한 개를 리턴하는 노드 클래스.
# 이 클래스의 evaluate 메소드는 초기화되었을 때의 값을 단순히 리턴한다
class paramnode:
    def __init__(self, idx):
        self.idx = idx
    
    def evaluate(self, inp):
        return inp[self.idx]
    
# 상수값을 리턴하는 노드 클래스.
# 이 클래스의 evaluate 메소드는 초기화되었을 때의 값을 단순히 리턴한다
class constnode:
    def __init__(self, v):
        self.v = v
    def evaluate(self, inp):
        return self.v
```

아래는 노드에 적용할 몇 가지 함수들이 나와있다.<br>
fwrapper를 사용하여 이름과 인자 개수를 함수에 전달하고있다.

```python
# lambda 는 간단한 인라인 함수라고 생각하면 된다.

# 함수 1 : 덧셈 
addw = fwrapper(lambda l:l[0] + l[1], 2, 'add')
# 함수 2 : 뺄셈 
subw = fwrapper(lambda l:l[0] - l[1], 2, 'subtract')
# 함수 3 : 곱셈 
mulw = fwrapper(lambda l:l[0] * l[1], 2, 'multiply')

def iffunc(l):
    if l[0] > 0:
        return l[1]
    else:
        return l[2]
# 함수 4 : 조건문 
ifw = fwrapper(iffunc, 3, 'if')

def isgreater(l):
    if l[0] > l[1]:
        return 1
    else:
        return 0
# 함수 5 : 대소 비교문
gtw = fwrapper(isgreater, 2, 'isgreater')

flist = [addw, mulw, ifw, gtw, subw]
```

### 위에서 만든 node 클래스를 사용하여 [그림 11-2] 와 같은 프로그램 트리를 만들어보자
우선, 아래는 [그림 11-2]와 이를 직접적으로 코드로 표현한것이다. 이것이 우리가 지금 만들려는 목표 프로그램이다.

## - 트리 제작 및 평가하기 -

위의 [그림 11-2]를 표현한 코드를 우리가 만든 트리를 이용하여 작성해보면 아래와 같다.

```python
def exampletree():
    return node(ifw, [
            node(gtw, [paramnode(0), constnode(3)]),
            node(addw, [paramnode(1), constnode(5)]),
            node(subw, [paramnode(1), constnode(2)]),
        ])
```

## - 프로그램 표시하기 -
이제 우리는 프로그램 트리를 수작업이 아닌 자동으로 생성할 것이다.<br>
<b>자동으로 생성되는 프로그램 트리의 결과를 쉽게 해석하기 위해</b> 출력문을 만들자.

우리가 짜놓은 프로그램 트리의 각 노드들은 자신의 역할을 나타내는 이름을 각각 가지고 있으므로 쉽게 만들 수 있다.

```python
from random import random, randint, choice
from copy import deepcopy
from math import log

# 함수 노드에서 사용할 삼수들에 대한 래퍼.
# 멤버 변수로 함수 이름, 함수 자신, 취할 인자의 개수를 가짐
class fwrapper:
    def __init__(self, function, childcount, name):
        self.function = function
        self.childcount = childcount
        self.name = name

# 자식 노드를 가진 함수 노드 클래스.
# fwrapper로 초기화된다. evaluate가 호출되면 자식 노드들을 평가한 후 그 결과로 이 함수를 호출한다.
class node:
    def __init__(self, fw, children):
        self.function = fw.function
        self.name = fw.name
        self.children = children
        
    def evaluate(self, inp):
        results = [n.evaluate(inp) for n in self.children]
        return self.function(results)
    
    # display 함수 1 : node 클래스에 트리의 문자열 표현을 보여줄 display 함수
    def display(self, indent = 0):
        print ('    '*indent) + self.name
        for c in self.children:
            c.display(indent + 1)
    
# 프로그램에 전달된 인자들 중 한 개를 리턴하는 노드 클래스.
# 이 클래스의 evaluate 메소드는 초기화되었을 때의 값을 단순히 리턴한다
# 어렵게 생각하지 말고 패러미터0, 패러미터1 과 같이 0, 1을 기록하는 패러미터 노드라고 할 수 있다.
class paramnode:
    def __init__(self, idx):
        self.idx = idx
    
    def evaluate(self, inp):
        return inp[self.idx]
    
    # display 함수 2 : 단순히 리턴할 인자의 인덱스를 출력하는 paramnode 클래스용 display 함수
    def display(self, indent = 0):
        print '%sp%d' % ('    '*indent, self.idx)
    
# 상수값을 리턴하는 노드 클래스.
# 이 클래스의 evaluate 메소드는 초기화되었을 때의 값을 단순히 리턴한다
class constnode:
    def __init__(self, v):
        self.v = v
        
    def evaluate(self, inp):
        return self.v
    
    # display 함수 3 : 마지막으로, constnode 클래스를 위한 display 함수
    def display(self, indent = 0):
        print '%s%d' % ('    '*indent, self.v)
```

바로 위에서 만든 display 함수들을 이용하여 트리를 출력해보겠다.

```python
def exampletree():
    return node(ifw, [
            node(gtw, [paramnode(0), constnode(3)]),
            node(addw, [paramnode(1), constnode(5)]),
            node(subw, [paramnode(1), constnode(2)]),
        ])
```

```python
exampletree = exampletree()
exampletree.display()
```

## < 3. 초기 개체군 만들기 >
초기 개체군은 보통 <b>무작위 프로그램 집합</b>으로 만든다.<br>
이는 <b>시작 절차를 간편하게</b> 하며, <b>초기 개체군이 다양성을 확보</b>하게끔 한다.<br>

<u>무작위 프로그램을 생성하는 작업</u>은 무작위로 연계된 함수를 가진 루트 노드를 만들고<br>
필요한 자식 노드들을 무작위로 원하는 만큼 만드는 것이다.

또한, 이 자식 노드들은 차례로 연계된 무작위 자식 노드들을 가진다.

트리로 동작하는 대부분의 함수와 같이 이 작업도 아래와 같이 재귀적으로 쉽게 정의할 수 있다.

```python
# pc : 트리가 입력으로 취할 인자의 개수
# maxdepth : 자동으로 만들어질 최대 트리의 깊이
# fpr : 새로 생성할 노드가 함수 노드일 확률
# ppr : 함수 노드가 아닌 경우 paramnode일 확률
def makerandomtree(pc, maxdepth=4, fpr=0.5, ppr=0.6):
    if random() < fpr and maxdepth > 0:
        f = choice(flist)
        children = [makerandomtree(pc, maxdepth-1, fpr, ppr) for i in range(f.childcount)]
        return node(f, children)
    elif random() < ppr:
        return paramnode(randint(0, pc-1))
    else:
        return constnode(randint(0, 10))
```

위의 <b>makerandomtree 함수</b>는 <u>무작위 함수를 가진 노드를 만들고</u> 이 함수에 <u>필요한 자식 노드의 개수가 얼마인지를 찾는다.</u><br>
<u>자식 노드가 필요할 경우</u> <u>새로운 노드를 생성하기 위해 자신을 호출</u>한다.<br>

이런 방식으로 이 함수가 <u>더 이상 자식 노드들을 필요로 하지 않을 경우</u>에만 <u>종료 브랜치</u>를 만들어가면서 전체 트리를 만든다.<br>
(자식 노드가 필요하지 않은 경우는 그 함수가 상수나 입력 변수를 리턴하는 경우에 해당된다)

maxdepth를 주어 아주 큰 트리를 얻는것을 막고있다.

이 함수를 테스트해보자.

```python
random1 = makerandomtree(2)
random1.display()

result1 = random1.evaluate([7, 1])
result2 = random1.evaluate([2, 4])

print ;
print ('result1 : %d\nresult2 : %d') % (result1, result2)
```

```python
random2 = makerandomtree(2)
random2.display()

result1 = random2.evaluate([5, 3])
result2 = random2.evaluate([5, 20])

print ;
print ('result1 : %d\nresult2 : %d') % (result1, result2)
```

정답을 말하자면 [표 11-1]의 매핑 함수는 아래와 같다.

```python
def hiddenfunction(x, y):
    return x**2 + 2*y + 3*x + 5

print hiddenfunction(26, 35)
print hiddenfunction(8, 24)
print hiddenfunction(20, 1)
print hiddenfunction(33, 11)
print hiddenfunction(37, 16)
```

<u>우리는 이미 정답인 매핑 함수를 알고 있으므로</u> 매핑 함수를 거쳐 옳은 결과를 만들어내는 <u>데이터 세트</u>를 만들 수 있다.<br>
이 데이터 세트는 <b>자동으로 생성되는 프로그램 트리를 검증하는데 사용</b>된다.

아래는 우리가 이미 알고있는 매핑 함수를 이용하여 <b>데이터 세트를 만들어 내는 함수</b>이다.

```python
def buildhiddenset():
    rows = []
    for i in range(200):
        x = randint(0, 40)
        y = randint(0, 40)
        rows.append([x, y, hiddenfunction(x, y)])
    return rows
```

자, 이제 데이터 세티를 만들어보고 그 결과를 확인해보자.

```python
hiddenset = buildhiddenset()
print hiddenset
```

<u>이런 매핑을 가지는 함수를 과연 성공적으로 생산해낼 수 있는가를 평가하기 위해</u></b> 
이렇게 데이터 세트를 만들고 있는 것이다.

## - 성공여부 측정하기 -
최적화를 사용할 때에는 항상 <u>얼마나 좋은 해답인지를 측정하는 방법의 필요성에 직면</u>하게 된다.<br>
이 경우에도 숫자 출력에 대해 프로그램을 검사하기 때문에 <b>해당 데이터 세트에 대한 정확한 해답에 얼마나 근접했는지</b>를<br>
확인하는 것이 프로그램을 테스트하는 가장 쉬운 방법이다.

아래는 해답 근접 여부에 대해 점수를 매기는 평가함수이다.

```python
# tree : 자동으로 생성된 프로그램 트리
# s : 정답 데이터 세트
def scorefunction(tree, s):
    dif = 0
    for data in s:
        v = tree.evaluate([data[0], data[1]])  # data[0] : x, data[1] : y
        dif += abs(v-data[2])  # data[2] : 결과
    return dif
```

이 함수는 <b>함수의 결과</b>를 계산하고 <b>데이터 세트 안의 모든 x, y, 결과값과 비교</b>한다.<br>
<u>0 값을 리턴한다면 생선된 프로그램이 모든 결과를 바르게 얻었음을 의미</u>한다.

scorefunction을 이용하여 이전에 만들었던 random1, random2를 평가해보자.

```python
print scorefunction(random1, hiddenset)
print scorefunction(random2, hiddenset)
```

이전에 자동으로 생성했던 프로그램 트리인 random1, random2의 평가 결과가 나왔다.

한두 번 만에 0에 근접한 프로그램 트리가 나오지 않을 것은 당연할 것이다.<br>
(만약, 한두 번 만에 0이 나오면.. 로또를 사도 될 정도로 운이 좋은 것이다)<br>

아무튼, <b>우리는 이제 프로그램 트리를 평가할 수 있는 검증 방법을 확보하게 되었다..!!</b><br>
<u>이 평가 함수는 다음세대를 구성할 프로그램의 선정에 요긴하게 사용될 것이다.</u>

간략화를 위해 두 번째 유형[그림 11-4]의 서브트리 변경 방식 돌연변이만 수행해 보자.<br>
이 동작은 아래의 mutate 함수에서 구현하고 있다.

```python
# [그림 11-4]의 방식으로 구현한 돌연변이 함수
# t : 트리
# pc : 패러미터 개수
# probchange : 돌연변이가 일어날 확률
def mutate(t, pc, probchange=0.1):
    if random() < probchange:
        return makerandomtree(pc)
    else:
        result = deepcopy(t)
    if isinstance(t, node):  # t 가 node 클래스 타입이면 true, 아니면 false
        result.children = [mutate(c, pc, probchange) for c in t.children]
    return result
```

이 mutate 함수는 트리의 상단에서부터 시작하여 해당 노드를 돌연변이 시킬지를 판단한다.<br>
해당 노드가 돌연변이되지 않는다면 자식 노드들에 대해 mutate를 호출한다.

<b>즉, 트리 전부가 바뀔수도 있고 아무것도 바뀌지 않을 수도 있다.</b>

## < 6. 교배하기 >
<u>프로그램을 수정하는 다른 방법</u>으로 <b>교배(crossover)</b> 또는 <b>번식(breeding)</b> 기법이 있다.<br>
<b>이것은 <u>성공적인 두 개의 프로그램</u>을 결합하여 새로운 프로그램을 만든다.</b>
이 때, 한 브랜치를 다른 브랜치로 교체하는 방법을 보통 사용한다.

[그림 11-5]를 보자.

```python
# [그림 11-5] 서브트리 변경 방식 돌연변이
from IPython.display import Image 
Image(filename='pic11-5.jpg', width='1000')
```

<b>교배</b>를 수행하는 함수는 <b>입력으로 두 트리를 받아 이 둘 모두를 하향 순회</b>한다.<br>
만일, 무작위로 선정된 경계에 다다르면 이 함수는 브랜치 중 하나가 두 번째 트리의 브랜치로 교체된 첫 번째 트리의 복사본을 리턴한다.

<b>양 트리를 동시에 순회하면 교배는 각 트리의 동일 레벨에서 적당히 발생할 것이다.</b>

아래는 교배를 수행하는 crossover 함수이다.

```python
def crossover(t1, t2, probswap=0.7, top=1):
    if random() < probswap and not top:
        return deepcopy(t2)
    else:
        result = deepcopy(t1)
        if isinstance(t1, node) and isinstance(t2, node):
            result.children = [crossover(c, choice(t2.children), probswap, 0) for c in t1.children]
        return result;
```

교배 함수를 만들었으니 테스트를 해보자.

```python
random1 = makerandomtree(2)
print scorefunction(random1, hiddenset)
random1.display()
```

```python
random2 = makerandomtree(2)
print scorefunction(random2, hiddenset)
random2.display()
```

```python
cross = crossover(random1, random2)
print scorefunction(cross, hiddenset)
cross.display()
```

교배 결과 random1 과 random2 가 서로 <u>부분적으로 결합</u>했음을 볼 수 있다.<br>
또한, 이런 </u>교배 결과를 평가함수로 측정하면 이전과는 다른 결과를 나타냄</u>도 볼 수 있다.

다시 강조하자면, <b>한두 번 만의 교배만으론 정답인 맵핑 함수를 구할 수는 없을 것</b>이다.<br>
(만약, 한두 번 만의 교배 결과를 평가함수로 측정했더니 0이 나오면 로또를 사러 가도 된다는 점을 염두해 두도록 하자)

<b>몇몇 교배는 해답을 개선할 것이며 다음 세대를 위해 계속 유지될 것이라는 점에 희망을 두도록 하자.</b>

## < 7. 환경 구축하기 >
우리는 지금 유전자 프로그래밍을 위한 세 함수를 만들어냈다.

1. scorefunction : 성공 여부를 측정하기위한 평가 함수
2. mutate : 돌연변이 함수
3. crossover : 교배 함수

앞서 본 [그림11-1]처럼 프로그램 트리가 진화할 경쟁 환경을 구축해보자.

```python
from IPython.display import Image 
Image(filename='pic11-1.jpg', width='1000')
```

가장 처음에 언급했던 절차와 마찬가지로

<strong><font size='3'>
1. 무작위 프로그램 집합을 생성
2. 프로그램 트리들을 평가하여 최적부터 최악까지 정렬
3. 최적인 프로그램 트리들을 돌연변이 및 교배
4. 공간이 모자라면 최악인 프로그램 트리들을 버림
5. 2~4를 중단조건에 도달할 때까지 계속 반복
</strong></font>

위와 같은 절차로 진행할 것이다.<br>

아래에 나와있는 evolve 함수는 이 절차를 간략하게 수행하는 함수이다.

```python
# scorefunction에 기반을 두어 프로그램 트리 리스트의 순서를 결정하는 함수를 리턴
# 즉, 평가 점수를 보고 최적인 순서로 프로그램 트리를 정렬하는 함수이다
def getrankfunction(dataset):
    # scores[*][0] 은 평가 점수를, scores[*][1]은 프로그램 트리를 나타낸다
    def rankfunction(population):
        scores = [(scorefunction(t, dataset), t) for t in population]
        scores.sort()
        return scores
    return rankfunction

# pc : 패러미터 개수
# popsize : 초기 개체군 크기
# rankfunction : 프로그램의 목록을 최적부터 최악까지 순위를 매기는데 사용하는 함수
# maxgen : 유전 알고리즘 최대 시행 횟수
# mutationrate : matate에 전달될 돌연변이 확률
# breedingrate : crossover에 전달될 교배 확률
# pexp : selectindex에 사용되며 이 값이 작을수록 낮은 인덱스를 획득, '적자생존'이 아닌 '적자+우연 생존'을 보장
# pnew : 돌연변이 및 교배를 시행할 확률, 새로운 혼합체의 생성을 보장
def evolve(pc, popsize, rankfunction, maxgen=500, mutationrate=0.1, breedingrate=0.4, pexp=0.7, pnew=0.05):
    # 낮은 숫자를 선호하도록 무작위 숫자를 리턴함
    # 이는 최적순서대로 프로그램 트리가 정렬될텐데 최적인 랜덤한 두개를 선택하도록 유도하기 위함임
    # 분모인 pexp가 작을수록 낮은 숫자를 얻을 가능성이 높음
    # 왜?
    # log(3/10) = log(3) - log(10) = 0.477 - 1 = -0.522 
    # log(7/10) = log(7) - log(10) = 0.845 - 1 = -0.154
    # 기본적으로 분모가 낮으면 분수값이 커지므로..!
    def selectindex():
        return int(log(random())/log(pexp))
    
    # 무작위로 초기 개체군을 생성함
    population = [makerandomtree(pc) for i in range(popsize)]
    for i in range(maxgen):
        scores = rankfunction(population)
        print scores[0][0]
        if scores[0][0] == 0: break
            
        # 두 개의 최고를 선택함
        newpop = [scores[0][1], scores[1][1]]
        
        # 다음 세대를 만듦
        while len(newpop) < popsize:
            if random() > pnew:
                newpop.append(mutate(
                        crossover(scores[selectindex()][1], scores[selectindex()][1], probswap = breedingrate),
                        pc,
                        probchange = mutationrate))
            else:
            # 무작위 노드를 섞어 넣음
                newpop.append(makerandomtree(pc))
        
        population = newpop
    scores[0][1].display()
    return scores[0][1]
```

이 evolve 함수는 다음과 같이 동작한다.

1. 초기에 무작위 개체군을 생성한다.
2. 최적부터 최악까지 프로그램에 순서를 매기는 rankfunction를 수행한다.
3. 최적 프로그램은 변경 없이 다음세대에 그대로 전달된다. 이 과정을 엘리티즘(elitism) 이라고 한다.
4. 다음세대의 나머지는 상위 순위 근처에 있는 프로그램들을 무작위로 선택하고 그들을 교배와 돌연변이시켜 구축한다.
5. 2~4 과정을 프로그램이 완벽한 점수인 0이나 maxgen에 도달할때까지 반복해서 수행한다.

<b>이제 수학 데이터 세트(x, y, 결과)를 바탕으로 이에 맞는 매핑 수학 함수식을 자동으로 찾아내는 프로그램을 만들 준비가 끝났다..!!</b><br>
이제 코드로 직접 실행해보자.

```python
# rankfunction
rf = getrankfunction(buildhiddenset())

evolve(2, 500, rf, mutationrate=0.2, breedingrate=0.1, pexp=0.7, pnew=0.1)
```

위에서 결과가 나왔는데 이는

x^2 + 6 + x + y + x + x - 1 + y = x^2 + 3x + 2y + 5

이며 <b>바로 우리가 찾던 매핑 수학 함수</b>를 유전자 프로그래밍을 통해 <b>올바르게 찾았다</b>는 것을 알 수 있다.<br>
(p_0는 x, p_1는 y임을 기억하자)

한번 더 유전자 프로그램을 돌려보겠다.

```python
# rankfunction
rf = getrankfunction(buildhiddenset())

evolve(2, 500, rf, mutationrate=0.2, breedingrate=0.1, pexp=0.7, pnew=0.1)
```

또 평가 결과가 0이 떠서 맞춘것으로 보인다. <u>그런데 좀 지저분</u>하다.<br>
그래도 마찬가지로 <u>식을 정리하면 같은 정답식을 도출</u>함을 알 수 있다.

이는 유전자 프로그래밍의 중요한 속성으로<br>
<b>찾은 해답이 바르거나 아주 좋을지라도 생성되는 방식에 의해 인간 프로그래머가 설계한 어떤 것보다 훨씬 더 복잡할 것</b>이다란 점이다.

아무것도 하지 않는 프로그램 영역이 존재할 가능성이 있고 매번 동일한 값을 리턴하는 복잡한 공식도 있을 수 있다.<br>
5를 더했다가 5를 빼고 다시 7을 더했다가 7을 빼는 과정이 있을 수도 있다.

이를 해결하는 방법은 <u>트리의 불필요한 부분을 간소화하거나 삭제하는 것</u>인데<br>
현재의 경우 손으로 직접 풀거나 가지치기 알고리즘(pruning algorithm)을 사용하면 될 것이다.

## - 중요한 다양성 -
evolve 함수에서 최적에서 최악까지 프로그램 트리의 순서를 정렬하고<br>
최적인 두 개의 프로그램 트리만 취해 돌연변이 혹은 교배를 시켰다.

<u>어느정도 최적인 것들만 남기지 뭐하려고 최악인것들까지 유지되도록 했을까?</u><br>
<b>바로 로컬 최적화(local maxima)에 빠지지 않게끔 하기 위해서</b>이다.

최고의 해답 몇 개만 선정하는 방식 즉, <b>최적인 것들만 남기는 방식</b>은 <br>
개체군이 극단적으로 동질화되게끔하므로 평가는 꽤 좋을테지만 비슷한 평가만 계속 유지되는 상태에 빠지게 될 것이다.<br>
즉, 어느정도의 좋은 평가만을 최종적으로 확보하고 그 이후로는 개선이 불가능한 <u>로컬 최적화에 빠지게 되는 것</u>이다.

이것은 적당히 좋은 해답이 아주 많이 결합된 최적 해답을 가지면 더 좋은 결과를 낸다는 것을 의미한다.

<u>evolve 함수는 로컬 최적화에 빠지지 않기 위해 선정 과정에서 <b>다양성</b>을 조절하는 여러 부가 인자들을 가지고 있다.</u>

pexp값을 줄이면 <b>'적자생존'</b>에서 <b>'적자와 우연 생존'</b>을 가능하게끔 하고<br>
pnew값을 늘리면 때때로 <b>완전히 새로운 프로그램을 생성시키는 것</b>을 허락하기도 한다.

이 두 값들 모두 진화 과정에서 <b>다양성 정도를 증가</b>시키지만 <br>
<u>아주 나쁜 프로그램은 종국에는 항상 제거되어야</u> 하므로 <b><u>적당한 값을 취하는것이 바람직하다.</u></b>

## < 유전자 프로그래밍의 한계 및 주의점[1] ><br>
<font size=3><strong>
1) 주어진 문제에 맞는 유전자 표현(representation) 정의하기</strong></font>

        유전자 프로그래밍에서 개체들은 실행 가능한 프로그래밍 코드의 트리로 표현되며 이것이 잘 정의되어야 한다.        
<font size=3><strong>        
2) 적합도 함수(fitness function)는 반드시 신중하고 정겨하게 정의되어야 한다</strong></font>

        좋은 해에 대해서는 더 높은 적합도를 가지게끔 하여야 한다.        
<font size=3><strong>        
3) 개체군 크기, 돌연변이와 교배의 발생 확률, 개체 선택의 강도 등의 패러미터를 신중이 고려하여 결정하여야 한다</strong></font>

        개체군의 크기가 너무 작다면..? 
            해공간(solution space)를 충분히 탐색하지 못해 좋은 결과 획득에 실패할 것.
            
        돌연변이(mutation)와 교배(crossover)의 발생 확률이 너무 크다면..?
            너무 많은 유전자 조작이 일어나 좋은 결과 획득에 실패할 것.
            
        개체선택(selection)이 적절하지 않다면..?
            좋은 해가 될 수 있는 스키마(schema)를 와해시켜 좋은 결과 획득에 실패할 것.
            
        이는 자연에서도 마찬가지로 젹용된다.
<font size=3><strong>        
4) 유전자 알고리즘은 'deceptive'한 문제를 푸는 적합도 함수를 다룰 때 어려움을 겪는다.</strong></font>

        'deceptive' 적합도 함수란 그 값이 점점 향상되어 전역 최적해(global optimum)가 있는 방향으로 나아가고 있는 듯 하지만
        실제로는 잘못된 방향으로 개체를 유도하는 함수를 일컫는다.
        
        1의 보수 법에서 0은 '00000000'과 '11111111'과 같이 2가지로 표현되는데
        이 문제를 풀기 위한 적합도 함수를 만드는것은 어렵다.
        
        그래도 이를 reverse연산을 통해 해결하는 논문[2]이 2008년도에 발표되었다.
<font size=3><strong>
5) 조기 수렴 문제에 빠질 수 있다.</strong></font>

        다른 개체들보다 굉장히 우수한 개체들이 초기에 나타났을 경우, 이 개체 혹은 같은 위치 및 성격을 갖는 개체들은 세대롤 거쳐
        계속 번식할 것이며 이는 전체 개체군(population)의 다양성 저하를 초래할 수 있다.
        
        즉, 지역 해에 조기수렴하게 될 가능성이 있다.
        
        이런 현상은 개체군의 크기가 작을 경우 일어날 가능성이 높다.
        
        이를 해결하기위한 방법으로 우리가 사용한 pexp값 처럼 우월한 개체들만 지나치게 선택하지 않게끔 하는 등의
        다양한 방법도 제시되고 있다.
<font size=3><strong>        
6) 마지막으로, 분석적으로 풀 수 있는 문제들은 유전자 알고리즘을 사용하지 않을 것이 권고된다.</strong></font>

        유전자 알고리즘으로도 문제를 충분히 풀 수 있겠지만 더 나은 방법이 있는데 굳이 사용할 필요가 없지 않는가.
        또한, 수학적으로 접근하는 방법이 가능하다면 이것만으로도 정확한 솔루션을 찾을 수 있다.
